Defining numeric and object columns

```python
num_cols = [
    'Rating', 'Total Reviews', 'Happy', 'Euphoric', 'Relaxed', 'Giggly',
    'Creative', 'Uplifted', 'Sleepy', 'Energetic', 'Aroused', 'Focused',
    'Talkative', 'Hungry', 'Tingly', 'Stress', 'Pain', 'Nausea',
    'Insomnia', 'Depression', 'Lack of Appetite', 'Muscle Spasms',
    'Seizures', 'Fatigue', 'Inflammation', 'Spasticity', 'Eye Pressure',
    'Cramps', 'Headaches', 'Dry Mouth', 'Dizzy', 'Anxious', 'Paranoid',
    'Headache'
]

obj_cols = [
    'Strain', 'Abr', 'Type', 'Description', 'Flavors',
    'Attribute Type', 'Parents'
]
```

Cleaning function - original iteration at bottom

```python
def clean(df):
    '''data cleaning function, removes unwanted characters'''
    df.Description = df.Description.apply(lambda x: str(x))#convert to string


   #iterbale columns to replace character
    char_col = ['Image Urls', 'Flavors', 'Attribute Type', 'Parents']


   #loop and replace
    for c in char_col:
        df[c] = df[c].apply(lambda x: x.replace("[", ""))
        df[c] = df[c].apply(lambda x: x.replace("]", ""))

   #more columns
    char_col = ['Description', 'Parents', 'Flavors',
                'Attribute Type', 'Strain']


   #loop and replace
    for c in char_col:
        df[c] = df[c].apply(lambda x: x.replace("'", ""))
  

   #this time, characters
    chars = ["‘", "’", "(", ")"]


   #loop and replace
    for c in chars:
        df.Description = df.Description.apply(lambda x: x.replace(c, ""))
  

   #random stragglers
    df['Description'] = df['Description'].apply(lambda x: x.replace('\xa0', ' '))
    df['Strain'] = df['Strain'].apply(lambda x: x.replace("’", ''))
    df[num_cols] = df[num_cols].apply(lambda x: x.replace('No Rank', 0))
    df[num_cols] = df[num_cols].apply(lambda x: x.replace('No Ranks', 0))
    #df = df.apply(lambda x: x.replace('[]', None))
    #df = df.drop(columns=("Attribute Type"))
  

    return df

#function call
clean(df)
```

Drop unused column

```python
df = df.apply(lambda x: x.replace('', 'N/A'))
df = df.drop(columns=("Attribute Type"))
```

SQL database table creation using strings as variable

```python
create_strain_table = '''
    CREATE TABLE strains (
      id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
      name TEXT,
      abrev TEXT,
      type TEXT,
      description TEXT,
      flavors TEXT,
      parents TEXT,
      rating REAL,
      reviews INT,
      happy INT,
      euphoric INT,
      relaxed INT,
      giggly INT,
      creative INT,
      uplifted INT,
      sleepy INT,
      energetic INT,
      aroused INT,
      focused INT,
      talkative INT,
      hungry INT,
      tingly INT,
      stress INT,
      pain INT,
      nausea INT,
      insomnia INT,
      depression INT,
      lackOfAppetite INT,
      muscleSpasms INT,
      seizures INT,
      fatigue INT,
      inflammation INT,
      spasticity INT,
      eyePressure INT,
      cramps INT,
      headaches INT,
      dryMouth INT,
      dizzy INT,
      anxious INT,
      paranoid INT,
      headache INT,
      imgUrl TEXT
);
  '''

#executing variable
sl_curs.execute(create_strain_table)


create_user_table = '''
  CREATE TABLE users (
    id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
    username VARCHAR NOT NULL UNIQUE,
    password VARCHAR NOT NULL,
    name VARCHAR NOT NULL,
    email VARCHAR NOT NULL UNIQUE,
    currentUsage VARCHAR,
    medicalQuery BOOLEAN,
    recreationalQuery BOOLEAN,
    medicalPurpose VARCHAR,
    goal VARCHAR,
    pastUser BOOLEAN,
    useFreq VARCHAR,
    intakeTime VARCHAR,
    happy INT,
    euphoric INT,
    relaxed INT,
    giggly INT,
    creative INT,
    uplifted INT,
    sleepy INT,
    energetic INT,
    aroused INT,
    focused INT,
    talkative INT,
    hungry INT,
    tingly INT,
    stress INT,
    pain INT,
    nausea INT,
    insomnia INT,
    depression INT,
    lackOfAppetite INT,
    muscleSpasms INT,
    seizures INT,
    fatigue INT,
    inflammation INT,
    spasticity INT,
    eyePressure INT,
    cramps INT,
    headaches INT
);

'''

#executing variable
sl_curs.execute(create_user_table)


create_reviews_table = '''
  CREATE TABLE reviews (
    id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
    helpfulness INT,
    flavor INT,
    intakeMethod VARCHAR,
    timeOfUse VARCHAR,
    goal VARCHAR,
    strain_id INT,
    user_id INT,
    notes TEXT,
    timestamp VARCHAR
);

'''


# executing variable
sl_curs.execute(create_reviews_table)
```

Get tuple of rows for insert statement

```python
strains = []
for strain in range(len(df)):
    strains.append(tuple(df.loc[strain]))
```

Loop to insert rows

```python
for strain in strains:
    insert_data = '''
      INSERT INTO strains
      (name, abrev, type, description, flavors,
       parents, rating, reviews, happy, euphoric,
       relaxed, giggly, creative, uplifted, sleepy,
       energetic, aroused, focused, talkative, hungry,
       tingly, stress, pain, nausea, insomnia,
       depression, lackOfAppetite, muscleSpasms, seizures,
       fatigue, inflammation, spasticity, eyePressure,
       cramps, headaches, dryMouth, dizzy, anxious,
       paranoid, headache, imgUrl)
      VALUES ''' + str(strain[:]) + ';'
    sl_curs.execute(insert_data)
```

Commit and close cursor/connenction

```python
conn.commit()
sl_curs.close()
conn.close()
```
